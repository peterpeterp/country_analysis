```python
###########
# check data availability
###########
tas=SEN.ensemble(['tas','CORDEX','rcp45'])
```

```python
tas_BC=SEN.ensemble(['tas','CORDEX_BC','rcp45'])
```

```python
EWEMBI_obs=SEN.selection(filters=['tas','EWEMBI'])
```

```python
##############
# Benefits of Bias-Correction
##############
tas_obs=EWEMBI_obs[0].period['ref'].copy()

fig, axes = plt.subplots(nrows=3, ncols=4,figsize=(12,6))
axes=axes.flatten()
count=0

lat=tas['IPSL'].lat.copy()
lon=tas['IPSL'].lon.copy()

for model in ['IPSL','ECEARTH','HADGEM2','MPIESM']:
    data=tas[model]
    to_plot=data.period['ref']-tas_obs
    im=plot_map(to_plot,lat,lon,
                title=data.model,
                ax=axes[count],
                color_range=[-1,1],
                color_bar=False,
                color_palette=plt.cm.RdBu
               )
    count+=1
    
for model in ['IPSL','ECEARTH','HADGEM2','MPIESM']:
    data=tas_BC[model]
    to_plot=data.period['ref']-tas_obs
    im=plot_map(to_plot,lat,lon,
                title=data.model,
                ax=axes[count],
                color_range=[-1,1],
                color_bar=False,
                color_palette=plt.cm.RdBu
               )
    count+=1
    
for i in range(4):
    axes[count+i].axis('off')

cbar_ax=fig.add_axes([0.3,0.2,0.4,0.6])
cbar_ax.axis('off')
cb=fig.colorbar(im,orientation='horizontal',label='Temperature Change [K]')
plt.show() 
```
